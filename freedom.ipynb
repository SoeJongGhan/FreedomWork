```python
import PyPDF2

def pdf_to_text(pdf_path, output_text_path):
  """
  Конвертирует PDF-файл в текстовый файл.

  Args:
    pdf_path: Путь к PDF-файлу.
    output_text_path: Путь для сохранения текстового файла.
  """

  with open(pdf_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    num_pages = len(pdf_reader.pages)

    with open(output_text_path, 'w', encoding='utf-8') as text_file:
      for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        text = page.extract_text()
        text_file.write(text)

# Загрузка файла в Google Colab
from google.colab import files
uploaded = files.upload()

# Получение имени загруженного файла
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

# Пути к файлам (замените на свои)
pdf_file = fn  # Имя загруженного PDF-файла
text_file = 'output.txt'

path="/content/резюме 7.pdf"


# Вызов функции для конвертации
pdf_to_text(path, text_file)

print(f"PDF успешно конвертирован в {text_file}")
```
